## Problem 1: Mobile Recharge Validation System

**Problem Statement:**
Create a function to validate a mobile recharge request.

**Rules:**
Recharge amount must be ≥ ₹50

Valid plans: 199, 299, 399, 599

User should be allowed to retry if invalid

```python
# list of valid recharge plans
valid_plans = [199, 299, 399, 599]

#function to validate recharge amount
def validate_recharge(amount):
    #check if amount is less than minimum required
    if amount < 50:
        return "Recharge amount must be at least 50 rupees."
    #check if amount is not in the list of valid plans
    elif amount not in valid_plans:
        return "Invalid plan selected."
    #if all conditions pass
    else:
        return "Recharge successful!"
#retry system using while loop
while True:
    try:
        #take user input and convert to integer
        recharge = int(input("Enter recharge amount: "))
        #call validation function
        result = validate_recharge(recharge)
        #print result message
        print(result)
        #exit loop if recharge is successful
        if result == "Recharge successful!":
            break
        else:
            print("Please try again.\n")

    #handle invalid input like text or symbols
    except ValueError:
        print("Please enter a valid number.\n")
```

## Problem 2: Inventory Reorder Alert System
**Problem Statement**
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise


```python
#function to check stock levels
def check_inventory(inventory):
    #loop through products and their stock
    for product, stock in inventory.items():
        
        #check if stock is below threshold
        if stock < 15:
            print(f"{product} : Reorder Alert")
        else:
            print(f"{product} : Stock OK")


#sample inventory data
inventory = {
    "Laptop": 8,
    "Mouse": 25,
    "Keyboard": 12
}

#call function
check_inventory(inventory)
```

## Problem 3: Student Result Processing System
**Problem Statement**
Create a function that:
Accepts a list of student marks

Calculates average marks

Prints Pass if average ≥ 50 else Fail


```python
#function to check student result
def check_res(marks):
    total = 0  # Initialize total marks

    #calculate total using loop
    for i in marks:
        total += i

    average = total / len(marks)  #calculate average

    #check pass or fail condition
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


#example list of marks
marks = [85, 44, 21]

#call function and print result
result = check_res(marks)
print("Result:", result)
```

## Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50

₹12 per km

Peak hour → 25% extra

Allow retry using while loop


```python
# Function to calculate cab fare
def calculate_fare(distance, peak_hour):

    # Base fare calculation
    fare = 50 + (12 * distance)

    # Add 25% extra during peak hours
    if peak_hour == "yes":
        fare = fare + (0.25 * fare)

    return fare


# Retry option using while loop
while True:

    # Take distance input
    distance = float(input("Enter distance travelled (in km): "))

    # Ask if it is peak hour
    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    # Calculate fare using function
    total_fare = calculate_fare(distance, peak_hour)

    # Display fare
    print("Total Fare: ₹", round(total_fare, 2))

    # Ask user if they want to retry
    retry = input("Do you want to calculate again? (yes/no): ").lower()

    if retry != "yes":
        break
```

## Problem 5: Employee Attendance Eligibility Checker
**Problem Statement**
Create a function that:

Accepts attendance list ("P", "A")

Calculates attendance percentage

Returns Eligible if ≥ 75% else Not Eligible


```python
# function to check attendance eligibility
def check_att(att):
    total = len(att)      # total days
    present = 0           # present count

    # count present days
    for d in att:
        if d == "P":
            present += 1

    # calculate percentage
    percent = (present / total) * 100

    # check eligibility
    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# attendance input
att = ["P", "P", "A", "P", "A", "P"]

# function call
result = check_att(att)

# print result
print(result)
```

## Problem 6: Password Strength Checker
**Problem Statement**
Create a function to check password strength.

**Rules:**
Minimum length = 8

Must contain at least one digit

Must contain at least one special character (@#$)


```python
# function to check password strength
def check_password(password):
    has_digit = False        # check for number
    has_special = False      # check for special character

    # minimum length check
    if len(password) < 8:
        return "Weak Password (minimum 8 characters required)"

    # loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    # final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (must include at least one digit and one special character @#$)"


# password input
password = input("enter password:")

# function call
result = check_password(password)

# display result
print(result)
```

# Problem 1: Unique Words in a Sentence
**Problem Statement**
Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves

**Input**
Enter sentence: python is easy and python is powerful

**Expected Output**
Unique words count: 5  
Unique words: {'python', 'is', 'easy', 'and', 'powerful'}


```python
#accepting input from user
sentence = input("Enter sentence:")
#converting to lowercase to avoid case-sensitivity
sentence = sentence.lower()
#using split function to convert sentence into individual words
words = sentence.split()
#converting words into a set to remove duplicates
unique = set(words)
#count of unique words 
count = len(unique)
print("Unique words count:", count)
#to print unique words
print("Unique words:", unique)
```

# Problem 2: Highest Salary from Employee Data
**Problem Statement**
Store employee names and their salaries, and determine which employee has the highest salary.

**Input Format**
A dictionary where:
Keys represent employee names
Values represent employee salaries
Example Input
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

**Expected Output**
Highest Salary: Ravi - 75000


```python
#given employee dictionary
employees = {
              "Ravi": 75000,
              "Anita": 68000,
              "Kiran": 72000
            }
#variables to assume highest salary and name
hname = " "
hsalary= 0
#loop through employees
for i in employees:
    #comparing to find highest salary and name
    if employees[i] > hsalary:
        hsalary = employees[i]
        hname = i
#printing result found
print("Highest Salary:",hname,"-",hsalary)
```

# Problem 3: Find Maximum and Minimum Values
**Problem Statement**
Find the maximum and minimum values from a list of numbers.

**Input Format**
A list of integers where each element represents a numeric value.
Example Input
numbers = [45, 22, 89, 10, 66]

**Expected Output**
List: [45, 22, 89, 10, 66]  
Max: 89  
Min: 10


```python
#given list of input numbers
numbers = [45, 22, 89, 10, 66]
#initilaising with first number 
maxval = numbers[0]
minval = numbers[0]
#loop through numbers to compare each element.
for i in numbers:
    if i > maxval:
        maxval = i
    if i < minval:
        minval = i
#print original list, max and min value
print("List:", numbers)
print("Max:", maxval)
print("Min:", minval)
```

# Problem 4: Count Products Above a Price Threshold
**Problem Statement**
Count how many products have a price greater than 1000.

**Input Format**
A list of integers where each element represents the price of a product.
Example Input
prices = [450, 1200, 899, 1500, 300]

**Expected Output**
Products above 1000: 2


```python
#given input
prices = [450, 1200, 899, 1500, 300]
#counter initialised to 0 
count = 0 
#loop to check each price above 1000
for i in prices:
    if i > 1000:
        count += 1 #counting above 1000
#printing result
print("Products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage
**Problem Statement**
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

**Example Input**
attendance = ["P", "P", "A", "P", "P"]

**Expected Output**
Attendance Percentage: 80.0


```python
#given input
attendance = ["P", "P", "A", "P", "P"]
#present counter initialised to 0 
p = 0
#loop to count present 
for i in attendance:
    if i == "P":
        p += 1 #counting
#percentage calculation
percent=(p/len(attendance))*100
#print result
print("Attendance Percentage:", percent)
```

# Problem 6: Remove Duplicate Phone Numbers
**Problem Statement**
You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.

**Input Format**
A list of integers where each integer represents a phone number.
Example Input
phone_numbers = [9876543210, 9123456789, 9876543210]

**Expected Output**
Unique phone numbers: {9876543210, 9123456789}


```python
#given input
phone_numbers = [9876543210, 9123456789, 9876543210]
# Convert list to set to remove duplicate values
unique = set(phone_numbers)
#print result
print("Unique phone numbers:", unique)
```

# Problem 7: Count Character Frequency
**Problem Statement**
Count how many times each character appears in a given string.
    
**Expected Output**
{'p': 2, 'y': 1, 't': 1, 'h': 1, 'o': 1, 'n': 1}



```python
#input text
text = "python"
# Empty dictionary to store frequency 
freq = {}  
# Count each character
for i in text:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
#print frequency
print(freq)
```

# Problem 10: Calculate Average Marks
**Problem Statement**
Calculate the average marks of a student from a list of marks.
    
**Expected Output**
Average Marks: 72.5


```python
#input list of marks
marks = [70, 75, 80, 65]
# Calculate average marks of student
avg = sum(marks) / len(marks)
#print result
print("Average Marks:",avg)
```
